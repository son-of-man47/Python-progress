# COMPARING THE TOTAL PROFITS OF THEE TWO TERRITORIES

```python
###creating an array for the territories

terri = []
a = np.array(df["COUNTRIES"])
for i in a:
    if i == "Nigeria" or i == "Ghana":
        terri.append("Anglophone")
    else:
        terri.append("Francophone")
terri
```

```python
#######converting it to a column

df = pd.concat([df, pd.DataFrame(terri)], axis = 1)
```

```python
# df.columns

######changing the name to territories
name = {0: "TERRITORIES"}
df.rename(columns=name, inplace=True)
df.head()
```

```python
####groupby and sum profits

terr = df.groupby("TERRITORIES")["PROFIT"].sum()
terr
```

```python
terr.plot(kind = "barh")
```

# COMPARE THE PROFITS OF EACH MONTH IN PERCENTAGE

```python
sen = df[(df.YEARS == 2019)]
yr = sen.groupby("MONTHS")["PROFIT"].sum()
y29 = yr.reset_index()
y29.columns = ["Months", "Total_profit"]
# y29

sum_prof = y29["Total_profit"].sum()

perc_prof = []
for i in np.array(y29["Total_profit"]):
    perc = round((i/sum_prof) * 100, 2)
    perc_prof.append(perc)
# print(perc_prof)


y29 = pd.concat([y29, pd.DataFrame(perc_prof)], axis = 1)
name = {0: "Percentage_profit"}
y29.rename(columns=name, inplace=True)
y29

y29.plot(x = "Months", y = "Percentage_profit", kind = "barh", color = "red", rot = 6)
```
