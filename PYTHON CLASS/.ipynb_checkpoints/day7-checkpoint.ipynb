```python
##printing the number of occurences of a value in the tz column

new_df["tz"].value_counts().head(10)  ##.value_counts() counts the number of occurances
```

```python
##turning it to a function

def col_count(col):
    c_count = new_df[col].value_counts().head(10)
    return c_count   
```

```python
type(col_count("a"))
```

```python
##this is a better way by first converting to a dictionary so that we could change the empty rows

def counter_col(col):
    dict_count = {}
    for i in np.array(new_df[col]):
        dict_count[i] = dict_count.get(i,0) + 1
    return dict_count
```

```python
counter_col("tz")
```

```python
keys = []
values =[]
for i, j in counter_col("tz").items():
    keys.append(i)
    values.append(j)
```
