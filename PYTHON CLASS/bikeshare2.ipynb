# function to call a particular dataset and split the start_time and end_time

```python
import pandas as pd

def get_city():
    city = input("enter city: ").lower()
    df = pd.read_csv(cities[city])
    return df

def split(df):
    df[["start_date", "start_time"]] = df["Start Time"].str.split(" ", expand = True)
    df[["end_date", "end_time"]] = df["End Time"].str.split(" ", expand = True)
    return df
```

```python
city = get_city()
new_city = split(city)
new_city
```

```python
new_city.info()
```

# renaming a column name

```python
name = {"Unnamed: 0": "id"}
new_city.rename(columns = name, inplace = True)
new_city.head()
```

# converting the start_date and enddate to date formats

```python
new_city["start_date"] = pd.to_datetime(new_city["start_date"])
new_city["end_date"] = pd.to_datetime(new_city["end_date"])
new_city.info()
```

# adding a day of the week column 

```python
from datetime import date
new_city["start_day"] = new_city["start_date"].dt.day_name()
new_city["end_day"] = new_city["end_date"].dt.day_name()
```

```python
new_city.head()
```

# adding a month column

```python
new_city["start_month"] = new_city["start_date"].dt.month_name()
new_city["end_month"] = new_city["end_date"].dt.month_name()
```

```python
new_city.head()
```

# adding a yr column

```python
new_city["start_yr"] = new_city["start_date"].dt.year
new_city["end_yr"] = new_city["end_date"].dt.year
```

```python
new_city.head()
```

# filling the na values with 'missing'

```python
new_city.head()
```

```python
###checking for colums with na values and giving them new names

nan = new_city.isna().sum().reset_index() ##the reset index gives the columns indexes so we can change them
nan.columns = ["variables", "count"]
nan.head()
```

```python
import numpy as np

missing = nan[nan["count"] > 0]
m_s=np.array(missing["variables"])
m_s
```

```python
for i in m_s:
    new_city[i] = new_city[i].fillna("missing")
```
