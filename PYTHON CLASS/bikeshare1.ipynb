# function to call a particular dataset and split the start_time and end_time

```python
import pandas as pd

def get_city():
    city = input("enter city: ").lower()
    df = pd.read_csv(cities[city])
    return df

def split(df):
    df[["start_date", "start_time"]] = df["Start Time"].str.split(" ", expand = True)
    df[["end_date", "end_time"]] = df["Start Time"].str.split(" ", expand = True)
    return df
```

```python
city = get_city()
new_city = split(city)
new_city.head()
```

# more splitting

```python
new_city[["start_yr", "start_month", "start_day"]] = new_city["start_date"].str.split("-", expand =True)
new_city[["end_yr", "end_month", "end_day"]] = new_city["end_date"].str.split("-", expand =True)
new_city
```

# renaming a column name

```python
name = {"Unnamed: 0":"id"}
new_city.rename(columns = name, inplace=True)
new_city.columns
new_city.head()
```

# converting the start_date and enddate to date formats

```python
new_city["start_date"] = pd.to_datetime(new_city["start_date"])
new_city["end_date"] = pd.to_datetime(new_city["end_date"])
new_city.info()
```
