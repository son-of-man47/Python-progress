```python
import pandas as pd


###putting the 3 file path in the folder into a dictionary for easy access
city_data = {
    "chicago": "C:/Users/IROANYA/Documents/DATA ANALYSIS FILES/PYTHON CLASS/bikeshare/chicago.csv",
    "new_york": "C:/Users/IROANYA/Documents/DATA ANALYSIS FILES/PYTHON CLASS/bikeshare/new_york_city.csv",
    "washington": "C:/Users/IROANYA/Documents/DATA ANALYSIS FILES/PYTHON CLASS/bikeshare/washington.csv"
}
```

```python
###function to request a city data

def get_city_data():
    city = input("enter city: ").lower()
    df = pd.read_csv(city_data[city])
    return df

##splitting a column

def clean(df):
    df[["start_date", "start_time"]] = df["Start Time"].str.split(" ", expand = True) ##splitting
    df[["End_date", "End_time"]] = df["End Time"].str.split(" ", expand = True)       ##splitting
    return df
```

```python
city = get_city_data()
new_city = clean(city)
new_city.head()
```

```python
new_city[["start_year", "start_month", "start_day"]] = new_city["start_date"].str.split("-", expand = True)
# new_city.head()
new_city[["end_year", "end_month", "end_day"]] = new_city["End_date"].str.split("-", expand = True)
new_city.head()
```

```python
new_city["start_date"] = pd.to_datetime(new_city["start_date"])     ##converting the date
new_city["End_date"] = pd.to_datetime(new_city["End_date"])          ###converting the date
```
