# CLEANING PROCESS

A function to check the na values in the dataframe

```python
def nan(df):
    nan = df.isna().sum().reset_index()
    nan.columns = ["columns", "counts"]
    nan
    missing = nan[nan["counts"] > 0]
    m_s =  np.array(missing["columns"])
    return m_s
```

Convering the orderdate column to a dtetime dtype
